# Finite-Difference Method for Diffusion

```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

$\frac{du}{dt} = \frac{d^2u}{dx^2}$\
$u(x, 0) = sin(\pi x)$\
$u(0, t) = u(1, t) = 0$\
$u_m^{(n)} = u(m\Delta x, n\Delta t)$\
$u_m^{(n+1)} = u_m^{(n)} + \frac{\Delta t}{\Delta x^2}(u_{m-1}^{(n)} + u_{m+1}^{(n)} - 2u_m^{(n)})$\

```python
dx = 0.05
dt = 0.0005
M = int(1/dx)
N = int(1/dt)
Mat = np.zeros((N+1, M+1))
# Mat[n][m] = u(m*dx, n*dt)

# Initial Condition
for i in range(M):
    Mat[0][i] = math.sin(math.pi*i*dx)
```

```python
# FDM Iteration
for n in range(N):
    for m in range(1,M):
        Mat[n+1][m] = Mat[n][m] + (dt/dx**2)*(Mat[n][m-1]+Mat[n][m+1]-2*Mat[n][m])
```

Analytic solution: $u(x, t) = e^{-\pi^2t}sin\pi x$

```python
# Analytic Solution
Mat_a = np.zeros((N+1, M+1))
for n in range(N+1):
    t = n*dt
    for m in range(M+1):
        x = m*dx
        Mat_a[n][m] = math.e**(-math.pi**2*t)*math.sin(math.pi*x)
```

```python
Timestamp = [n for n in range(0,N+1,200)]
X = np.array([m*dx for m in range(M+1)])

for i in Timestamp:
    print('T =', round(i*dt,1))
    plt.figure()
    plt.plot(X, Mat[i], label='FDM')
    plt.plot(X, Mat_a[i], linestyle='--', label='Analytic')
    plt.legend()
    plt.show()
```

```python
# Error Plot
for i in Timestamp:
    print('T =', round(i*dt,1))
    plt.figure()
    plt.plot(X, abs(Mat[i]-Mat_a[i]), label='Error')
    plt.legend()
    plt.show()
```
