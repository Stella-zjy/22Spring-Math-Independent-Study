## Finite Difference Method

```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

$\frac{du}{dt} = -2u$, $u(0)=1$

$\frac{u_{k+1} - u_{k}}{\Delta t} = -2u_k$, $u_{k+1} = (1-2\Delta t)u_k$

```python
dt = 0.005
L = 10
N = int(L//dt)+1

u = [1 for i in range(N+1)]
v = [math.e**(-2*dt*i) for i in range(N+1)]

for i in range(N):
    u[i+1] = (1-2*dt)*u[i]
```

```python
X = [i*dt for i in range(N+1)]
plt.figure()
plt.plot(X, u, label='FD Solution')
plt.plot(X, v, linestyle='--', label='Analytic Solution')
plt.legend()
plt.show()
```

```python
plt.figure()
plt.plot(X, abs(np.array(u)-np.array(v)), label='Error')
plt.legend()
plt.show()
```
